In the below code, I filter out the dataframe to contain only rows which are in the transportation industry of the economic sector. I named this new dataframe "new".

```python
e_sector = ["Transportation"]
new = df[df["Economic Sector"].isin(e_sector)]
```

This script below further filters out the data by finding the colunns worthy of further analysis in my project. It works by initializing a significant columns array, "sig_cols", then iterates through each column, checking to see if the dataframe contains very few values or very many values for that variable. In this script, I chose that the column must have at least 2 different values occur or no more than five. The lower bound is because some columns had the same value for each row in the entire dataframe, so I wanted to remove these columns since they did not offer much new information across this data set. I also restricted the upperbound to 5 so that anaylsis could be conducted across categorical variables with out having too many values to dilute the data set.

```python


sig_cols = []
for col in new.columns:
    vals = {}
    for i in range(len(new)):
        val = new.iloc[i][col]
        if val in vals:
            vals[val] += 1
        else:
            vals[val] = 1
    if 2 <= len(vals) <= 5:
        sig_cols.append(col)
    
print(sig_cols)
```

The final dataframe established in this program is "finished", which is the "new" dataframe restricted to the significant columns.

```python
finished = new[sig_cols]
```

The new dataframe is saved to a csv file.

```python
finished.to_csv("refined.csv")
```
